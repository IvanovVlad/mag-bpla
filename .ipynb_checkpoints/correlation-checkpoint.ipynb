```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_images(images):
    n = len(images)
    f = plt.figure()
    for i in range(n):
        f.add_subplot(1, n, i + 1)
        plt.imshow(images[i], cmap="gray")
    plt.show(block=True)
```

```python
satellite_img = cv2.imread("./img/Amap_pirogovo_50_85.png", cv2.IMREAD_GRAYSCALE)
satellite_img_copy = (np.copy(satellite_img))
satellite_img_copy[500:750, 500:750] = 0
crop_img = satellite_img[500:750, 500:750]
show_images([satellite_img, satellite_img_copy, crop_img])
del satellite_img_copy
```

```python
methods = [cv2.TM_CCOEFF, cv2.TM_CCOEFF_NORMED, cv2.TM_CCORR, cv2.TM_CCORR_NORMED, cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]
results = []
for method in methods:
    img = np.copy(satellite_img)
    res = cv2.matchTemplate(img, crop_img, method)
    results.append(res)
show_images(results)
```
